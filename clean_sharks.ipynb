```python
import pandas as pd
import numpy as np
from datetime import datetime as dt
```

```python
file_to_load = "shark_attacks.csv"
df = pd.read_csv(file_to_load)
df.head()
```

```python
#Drop data before 1975
df_1975 = df[df["Year"]>=1975]
df_1975.head()
```

```python
#Case Number is a better date indicator than other date column, the following code is meant to take the Case Number
# create a new column out of it and convert it to a date format
# first bit of code is to remove dates that have no month/day, just year
df_1975['Attack Date'] = df_1975["Case Number"].str.contains("00.00")==False
df_1975_clean = df_1975[df_1975['Attack Date']==True]

#this removes the extra digits on the case number, i.e. if multiple happened one day they are appended with a suffix of a,b,c
df_1975_clean['Attack Date'] = df_1975_clean['Case Number'].str[:10]
df_1975_clean
```

```python
#replacing the periods with hyphens, hoping this can be used to create a date format later
df_1975_clean['Attack Date'] = df_1975_clean['Attack Date'].str.replace('.','-')
df_1975_clean
```

```python
#create a new dataframe with only the important columns ##NOTE. this could probably be done earlier.
shark_trim = df_1975_clean[["Attack Date","Year","Type","Country","Area","Location","Activity","Sex ","Age","Injury","Fatal (Y/N)","Species "]]
shark_trim
```

```python
#Simple replace for the Type
shark_trim["Type"] = shark_trim.loc[:,"Type"].replace(
    {"Boat": "Boating",
     "Boatomg": "Boating",
     "Questionable":"Unknown",
     "Invalid":"Unknown"})
shark_trim["Type"].value_counts()
```
